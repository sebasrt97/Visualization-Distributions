## 1.7 Convert the data to the International System of Units (SI) and plot

Create a unique double plot for Weight and Height with 'Gender' as 'hue'

```python
df_weight["Height_m"]= df_weight["Height"]*2.54
df_weight["Weight_kg"]= df_weight["Weight"]*0.5435592

# Crear una figura con dos subgrficos (1 fila, 2 columnas)
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Grfico 1: Distribucin del Peso
sns.histplot(data=df_weight, x='Weight_kg', hue='Gender', ax=axes[0], palette='magma')
axes[0].set_title('Distribucin de Peso (SI)')
axes[0].set_xlabel('Peso (kg)')

# Grfico 2: Distribucin de la Altura
sns.histplot(data=df_weight, x='Height_cm', hue='Gender', ax=axes[1], palette='magma')
axes[1].set_title('Distribucin de Altura (SI)')
axes[1].set_xlabel('Altura (m)')

plt.tight_layout()
plt.show()
```

## 1.8 Plot the same boxplot data with Seaborn

Which one do you think is better and why?

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Boxplot para el Peso
sns.boxplot(data=df_weight, x='Gender', y='Weight_kg', hue='Gender', ax=axes[0], palette='Set2')
axes[0].set_title('Distribucin de Peso (kg)')
axes[0].set_ylabel('Peso (kg)')
axes[0].set_xlabel('Gnero')

# Boxplot para la Altura
sns.boxplot(data=df_weight, x='Gender', y='Height_m', hue='Gender', ax=axes[1], palette='Set2')
axes[1].set_title('Distribucin de Altura (m)')
axes[1].set_ylabel('Altura (m)')
axes[1].set_xlabel('Gnero')

# Ajustar diseo
plt.tight_layout()
plt.show()
```

## 1.9 Is there any outlier in height or weight?

Create figure with 2 subplots side by side

```python
fig, axes = plt.subplots(1,2, figsize=(12,6))

sns.boxplot(y=df_weight["Weight_kg"], ax=axes[0])
axes[0].set_title=("Outliners de peso:")
axes[0].set_ylabel("Peso KG")

sns.boxplot(y=df_weight["Height_cm"], ax=axes[1])
axes[1].set_title=("Outliners de peso:")
axes[1].set_ylabel("Altura M")

plt.tight_layout()
plt.show()
```
