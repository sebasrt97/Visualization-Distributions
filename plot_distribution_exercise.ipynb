## 1.2 Plot the same histogram data with Seaborn

```python
plt.hist(df_titanic['Age'].dropna(), bins=30)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution")
plt.show()
```

```python
prueba = df_titanic["Age"].dropna()

sns.displot(x=prueba, bins=20)
plt.ylabel("Frecuencia") # Cambiamos la etiqueta de Matplotlib
plt.show()
```

## 1.3 Plot the same data with Plotly library

Render in browser

```python
plt.hist(df_titanic['Age'].dropna(), bins=30)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution")
plt.show()
```

```python
import plotly.express as px

fig = px.histogram(
    df_titanic,
    x="Age",
    nbins=30,
    title="Age Distribution"
)

fig.update_layout(
    xaxis_title = "Age",
    yaxis_title="Frequency"
)

fig.show()
```
